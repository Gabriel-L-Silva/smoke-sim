# Interpolação Linear
A interpolação linear é a mais simples, é necessário montar um grid e usaremos os pontos disponíveis da seguinte forma:

\begin{equation}
y(x) = y_{1} + (x - x_{1})\frac{(y_{2}-y_{1})}{(x_{2}-x{1})}
\end{equation}

Interpolação com os índices - central sampling:
\begin{equation}
(i,j,k) = \left(\frac{x-x_{0}}{\Delta x}, \frac{y-y_{0}}{\Delta y}, \frac{z-z_{0}}{\Delta z}\right)
\end{equation}


```python
def linear(x, y):
    '''Interpolação linear para qualquer função, sem grid'''
    def R(P):
        def r(p):
            if(np.sum(x == p) == 1):
                return y[x == p][0]
            # Índice do ponto a esquerda de x0
            ix1 = np.sum(x < p)-1
            x1 = x[ix1]
            y1 = y[ix1]
            # Índice do ponto a direita de x0
            ix2 = np.sum(x < p)
            x2 = x[ix2]
            y2 = y[ix2]        
            yp = y1 + (p-x1)*(y2-y1)/(x2-x1)
            return yp
        if(type(P) == int or type(P) == float):
            return r(P)
        return np.array([r(p) for p in P])
    return R
```

```python
x = np.array([-1, 0, 1, 2]) # situação mais simples
y = np.array([5, 2, 10, 4])

interplinear = linear(x,y)
interplinearscipy = interp1d(x, y, kind="linear")

n = 100
p = np.linspace(-1,2,n)
py = interplinear(p)
pyscipy = interplinearscipy(p)

plt.figure(figsize=(12,6))
plt.plot(p,py, linewidth=4, color="red") # interpolação implementada
plt.plot(p,pyscipy,linewidth=1, color="yellow") # função do scipy
plt.plot(x, y, "o", color="black")
plt.show()
```

## Interpolação Linear - grid

```python
def interpGrid(f0, f1, t, h): #x0 = 0 #considerando uma celula de tamanho 1
    return f0 + (t)*(f1-f0)/h
    
def callLinearGrid(g,h,p):
    '''Interpolação linear - ideia de grid'''
    x0 = int(np.floor(p)) 
    x1 = x0 + 1
    f0 = g[x0]
    f1 = g[x1]
    p = p - np.floor(p)
    f = interpGrid(f0,f1,p, h)
    return f
```

```python
h = 1 #quando h = 1 - python =  minha função 
x = np.arange(0,10,h) # situação mais simples
y = x**2
print(x)
print(y)
p = 7 # na pos p a função é f

interplinear = linear(x,y)
interplinearscipy = interp1d(x, y, kind="linear")
#Passando valor do x
py = interplinear(p)
print("linear função " +  str(py))
pyscipy = interplinearscipy(p)
print("função pronta " +  str(pyscipy))

# Passando índice
pgrid = callLinearGrid(y, h, p)
print("função p/ o grid: " +  str(pgrid))
```

# Interpolação bilinear

Generalização da interpolação linear para uma função de 2 variáveis. 
É feita em 2 dimensões mas é linear em cada uma delas.

- 2 interpolações em x e 1 em y

```python
x = np.arange(-5,5,1)
y = np.arange(-5,5,1)
grid = np.meshgrid(x, y)
X = grid[0]
Y = grid[1]
Z = X**2 + Y**2 #z é um grid -  pontos dessa função no grid


fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.set_title("Pontos a serem interpolados")
ax.scatter(X, Y, Z)
```

```python
def callBilinear(g, hx, hy, px, py):
    x0 = int(np.floor(px))
    y0 = int(np.floor(py))
    x1 = x0 + 1
    y1 = y0 + 1
    px = px - np.floor(px) # mapeando px e py para um intervalo [0:1]
    py = py - np.floor(py)
    z0 = interpGrid(g[x0,y0], g[x1,y0], px, hx)
    z1 =  interpGrid(g[x0,y1], g[x1,y1], px, hx)
    z = interpGrid(z0, z1, py, hy)
    return z
```

```python
px = 5.7
py = 4.3
hx = 1 
hy = 1
f = callBilinear(Z, hx, hy, px, py)
print(Z)
print("bilinear " + str(f))
```

 # Interpolação de Catmull-Rom
 Separar vizinhos da interpolação em si.
 Queremos encontrar os coeficientes da equação $f = a_{3}t^3 + a_{2}t^2 + a_{1}t+ a_{0}$, na qual t é uma variável parametrizada que vai de 0 a 1.
 
 $a_{0} = f_{1}$
 
 Derivando $f(t)$, temos:
 
 $f'(t) =  3a_{3}t^2 +2a_{2}t + a_{1}$
 
 Podemos chamar $f'(t)$ de $d'(t)$.
 
 $d(t) = f'(t) =  3a_{3}t^2 +2a_{2}t + a_{1}$
 
 Agora tomaremos $d(t)$ em 0 e 1, ou seja, no ínicio e no final do intervalo:
 

 $d(0) = d_{1} = \frac{f_{2}-f{0}}{2} $
 
 $d(1) = d_{2} = \frac{f_{3}-f{1}}{2}$
 
 Se fixarmos t = 1 e substituirmos nas equações lineares, temos
 $f_{2} = a_{3} + a_{2} + a_{1} + a_{0}$
 E por fim:
 $d_{2} = 3a_{3} + 2a_{2}$ + a_{1}
 
Sendo assim o algoritmo da interpolação é:

 $d_{1} = \frac{f_{2}-f{0}}{2} $

 $d_{2} = \frac{f_{3}-f{1}}{2}$
 
 $D_{1} = d_{2} = \frac{f_{3}-f{1}}{2}$
 
 $a_{3} = d_{1} + d_{2} - 2D_{1}$
 
 $a_{2} = 3D_{1} - 2d_{1}-d_{2}$ 
 
 $a_{1} = d_{1}$
 
 $a_{0} = f_{1}$
 
 $r = a_{3}*f^3 + a2* f^2 + a_{1}*f + a_{0}$

```python
def normalize(x1, x2, p):
    return (p-x1)/(x2-x1)

def catmullrom(x, y):
    f0 = y[0]
    f1 = y[1]
    f2 = y[2]
    f3 = y[3]
    
    d1 = (f2 - f0) / 2 
    d2 = (f3 - f1) / 2 

    D = f2 - f1 

    a3 = d1 + d2 - 2 * D
    a2 = 3 * D - 2 * d1 - d2
    a1 = d1
    a0 = f1

    def r(t):
#         if(t < x[1] or t > x[2]):
#             print("O ponto está fora do intervalo1")
#             return
        p = normalize(x[1], x[2], t)
        return (a3*p**3 + a2*p**2 + a1*p + a0)
    
    return r
```

```python
#se o x for o mesmo, mesma coluna, se o y for o mesmo, mesma linha
def callcatmullrom(x, y):
    def R(T):
        def r(t):
            if(np.sum(x==t) == 1):
                return y[x == t][0]
            ix1 = np.sum(x < t) - 1
            ix2 = np.sum(x < t)
            n = len(y) - 1
            # os 4 pontos usados na interpolação
            if(ix1 == 0):
                ptsX = np.array([x[0], x[0], x[1], x[2]])
                ptsY = np.array([y[0], y[0], y[1], y[2]])
            elif (ix1 == n-1):
                ptsX = np.array([x[n-2], x[n-1], x[n], x[n]])
                ptsY = np.array([y[n-2], y[n-1], y[n], y[n]])
            else:
                ptsX = x[(ix1-1):(ix1+3)]
                ptsY = y[(ix1-1):(ix1+3)]
            f = catmullrom(ptsX,ptsY)
            return f(t)
        if(type(T) == int or type(T) == float):
            return r(T)
        return np.array([r(t) for t in T])
    return R
    
```

```python
x = np.array([-2, -1, 2, 5])
y = np.array([5, 2, 10, 4])
#y = x**2

r = catmullrom(x, y)
t = np.arange(x[1],x[2]+0.01,0.01)
f = np.array([r(T) for T in t])

plt.figure(figsize=(12,4))
plt.plot(x, y, "o")
plt.plot(t, f)
#plt.plot(t, t**2)
plt.title("Catmull-Rom sem as condições para os pontos de borda", fontsize=16)
plt.show()
#passa por todos os pontos, é interpolante FONTE: https://www.mvps.org/directx/articles/catmull/
```

# Catmull-Rom -  grid

```python
def get_neighbors(l, p):
    '''pegando os vizinhos'''
    if np.floor(p) == 0:
        f0 = 0
        f1 = 0
        f2 = 1
        f3 = 2
    elif np.floor(p) == 1:
        f0 = 0
        f1 = 1
        f2 = 1
        f3 = 2
    elif np.floor(p) == len(l) - 2:  #np.floor(p) == 1 ---- mesmo caso
        f0 = len(l) - 3
        f1 = len(l) - 2
        f2 = len(l) - 2
        f3 = len(l) - 1
    elif np.floor(p) == len(l) - 1: #np.floor(p) == 0 ---- mesmo caso 
        f0 = len(l) - 1
        f1 = len(l) - 1
        f2 = len(l) - 2
        f3 = len(l) - 3
    else:  
        f1 = int(np.floor(p)) 
        f0 = f1 - 1
        f2 = f1 + 1
        f3 = f1 + 2
        
    return l[f0], l[f1], l[f2], l[f3]
```

```python
def normalizeCR(p):
    '''normalizando o ponto'''
    return p - np.floor(p) #onde está o ponto em relação ao intervalo considerado

def catmullRomGrid(f,f0, f1, f2, f3):
    
    d1 = (f2 - f0) / 2 
    d2 = (f3 - f1) / 2 

    D = f2 - f1 

    a3 = d1 + d2 - 2 * D
    a2 = 3 * D - 2 * d1 - d2
    a1 = d1
    a0 = f1

    return a3*f**3 + a2*f**2 + a1*f + a0
    

def callCatmullRomGrid(g, p):
    '''interpolação em 1D'''
    f0, f1, f2, f3 = get_neighbors(g, p) #p é posição

    f = normalizeCR(p)
    
    return catmullRomGrid(f, f0, f1, f2, f3) 
   
```

```python
p = 5.9 # 7.5 entre 343 e 512
x = np.arange(0,10,1)
y = x**3
print(x)
print(y)
c = callCatmullRomGrid(y, p) 
print("Catmull-Rom: ", c)

plt.plot(x, y, label ="Curva teórica")
plt.plot([3.2,7.4,5.9], [32.864,405.272, 205.307], 'o', label="pontos interpolados")
plt.title("Catmull-Rom 1D para grid")
plt.legend()
```

```python
x = np.arange(-5,5,1)
y = np.arange(-5,5,1)
grid = np.meshgrid(x, y)
X = grid[0]
Y = grid[1]
Z = X**2 + Y**2 #z é um grid -  pontos dessa função no grid
```

```python
def callCatmullRom2D(grid, px, py):
    #pegando os vizinhos em 2D
    if(px == len(grid[0]) - 2):
#         print("p: ",np.floor(p))
        print("len ",len(grid[0]) )
        f0 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-3, :], px)
        f1 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-2, :], px)
        f2 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-2, :], px)
        f3 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
#     elif(px + 1 == len(grid[0]) - 2):
#         f0 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
#         f1 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
#         f2 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-2, :], px)
#         f3 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-3, :], px)
    elif(px + 1 > len(grid[0]) - 1):
        f0 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
        f1 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
        f2 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-2, :], px)
        f3 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-3, :], px)
    else:
        f0 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))-1, :], px)
        f1 = callCatmullRomGrid(grid[int(np.floor(py + 0.5)), :], px)
        f2 = callCatmullRomGrid(grid[int(np.floor(py + 0.5))+1, :], px)
        f3 = callCatmullRomGrid(grid[int(np.floor(py+ 0.5 ))+2, :], px)
        
    f = normalizeCR(py) 
    
    catmullRomGrid(f,f0, f1, f2, f3)

    return  catmullRomGrid(f,f0, f1, f2, f3)
```
